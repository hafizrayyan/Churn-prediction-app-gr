```python
label_encoder = LabelEncoder()

df["Gender"] = label_encoder.fit_transform(df["Gender"])
```

```python
oh = OneHotEncoder(sparse_output=False)
```

```python
geo_encoded = oh.fit_transform(df[['Geography']])
```

```python
geo_encoded.shape
```

```python
oh.get_feature_names_out()
```

```python
geo_df = pd.DataFrame(geo_encoded ,columns = oh.get_feature_names_out())
```

```python
df = pd.concat([df,geo_df],axis = 1)
```

```python
df
```

```python
df.columns
```

```python
X = df.drop(['RowNumber','CustomerId','Surname','Geography','Exited'],axis = 1)
y= df['Exited']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lr_param_grid= {
        'C' : [0.01,0.1,1,10,5] ,
        'penalty': ['l1','l2'] ,
        'solver' : ['liblinear'],
        'class_weight':[None, 'balanced']

}
```

```python
lr_grid = GridSearchCV(
    estimator = LogisticRegression(max_iter=1000),
    param_grid = lr_param_grid,
    scoring = 'recall'

)
```

```python
lr_grid.fit(X_train_scaled,y_train)
```

```python
d_tree = DecisionTreeClassifier(random_state= 42)
```

```python
d_param_grid = {
    'max_depth' : [None , 5 , 10 ,15, 20],
    'min_samples_split' : [2,5,8,10] ,
    'min_samples_leaf' : [1,2,4,6] ,
    'criterion' : ['gini' , 'endtropy'] ,
    'class_weight' : [None,'balanced']
}
```

```python
d_grid = GridSearchCV(
    d_tree ,
    d_param_grid ,
    scoring = "recall" ,
)

d_grid.fit(X_train,y_train)
```

```python
r_forest = RandomForestClassifier(random_state=42)
```

```python
rf_param_grid = {
    "n_estimators" : [10,20] ,
    "max_depth" : [None , 5,10] ,
    "min_samples_split" : [2,5,],
    "min_samples_leaf" : [2,4] ,
    "max_features" : ["sqrt","log2"] ,
    "class_weight" : [None,'balanced']
}
```

```python
rf_grid = GridSearchCV(
    r_forest,
    rf_param_grid ,
    scoring = 'recall',
    cv=3,
    n_jobs=-1,
    verbose=2
)
rf_grid.fit(X_train,y_train)
```

```python
gr = GradientBoostingClassifier()
```

```python
gr_param_grid={
    'learning_rate' : [0.1,0.15,0.25,] ,
    'n_estimators' : [300,400]


}
```

```python
gr_grid = GridSearchCV(
    gr,
    gr_param_grid ,
    scoring = 'recall',


)
gr_grid.fit(X_train,y_train)
```

```python
def evaluate_model(model,X_test,y_test,model_name) :
  y_pred = model.predict(X_test)

  print(f"\n{model_name}")
  print("Accuracy:", accuracy_score(y_test,y_pred))
  print(classification_report(y_test,y_pred))
```

```python
evaluate_model(lr_grid.best_estimator_,X_test,y_test,"Logistic Regression")
evaluate_model(d_grid.best_estimator_,X_test,y_test,"Decion Tree")
evaluate_model(rf_grid.best_estimator_,X_test,y_test,"Random Forest")
evaluate_model(gr_grid.best_estimator_,X_test,y_test,"Gradient Boosting")
```
